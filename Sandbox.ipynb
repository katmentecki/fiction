```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("lavin_results/meanshift_all_no_seeding.csv")
del df['Unnamed: 0']
```

```python
df_sorted = df.sort_values(by="group_label")
```

```python
big_genres = pd.read_csv("meta/datadictionary.csv")
gen_dict = {}
for i in big_genres.itertuples():
    gen_dict[i[1]] = i[2]
```

```python
genres_piped = list(df_sorted["genres"])
genres_main = []
for i in genres_piped:
    gen = i.split(" | ")
    g = []
    for z in gen:
        if "lavin" in z:
            g = [z]
        elif z != "teamred" and z!= "teamblack" and z!= "stew" and "random" not in z:
            #look up and append big genre
            try:
                g.append(gen_dict[z])
            except: 
                pass
        
    #merge duplicates
    g = list(set(g))
    if len(g) > 1:
        g = ["multi",]
    if len(g) == 0:
        g = ["non_genre"]
    genres_main.append(g)
df_sorted["final_genre"] = [i[0] for i in genres_main]
```

```python
df_sorted
```

```python
ratios_all_buckets = []
for bucket in set(list(df_sorted["group_label"])):
    matches = df_sorted.loc[df_sorted['group_label'] == bucket]
    ratios = Counter(list(matches["final_genre"]))
    ratios_all_buckets.append(ratios)
```

```python
percs = []
for m in ratios_all_buckets:
    p = Counter(dict([(i, m[i] / float(sum(m.values())) * 100.0) for i in m]))
    percs.append(p)
```

```python
for z in percs:
    #check grou
    print(z.most_common(3))
```
